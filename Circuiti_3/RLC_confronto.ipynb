```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'grid', 'no-latex'])
```

```python
V_res_1, V_l_1, fase_res_mod_1, fase_l_mod_1, freqs_1 = [],[],[],[],[]
with open('RLC_R_997.tmp.data', 'r') as f:
    data = f.readlines()
    for i in data:
        a,b,c,d,e  = i.strip('\n').split(' ')
        V_res_1.append(float(a))
        V_l_1.append(float(b))
        fase_res_mod_1.append(float(c))
        fase_l_mod_1.append(float(d))
        freqs_1.append(float(e))
```

```python
V_res_2, V_l_2, fase_res_mod_2, fase_l_mod_2, freqs_2 = [],[],[],[],[]
with open('RLC_R_4021.tmp.data', 'r') as f:
    data = f.readlines()
    for i in data:
        a,b,c,d,e  = i.strip('\n').split(' ')
        V_res_2.append(float(a))
        V_l_2.append(float(b))
        fase_res_mod_2.append(float(c))
        fase_l_mod_2.append(float(d))
        freqs_2.append(float(e))
```

```python
plt.subplots(1,2 , figsize=(10,5), dpi=200)
plt.subplot(1,2,1)
plt.title('Modulo')
plt.plot(freqs_1, V_res_1, 'o--',label='$R = 997 \Omega$', color='r', lw=1, ms=2)
plt.plot(freqs_2, V_res_2, 'o--', label='$R = 4021 \Omega$', color='b', lw=1, ms=2)
plt.xlabel('Frequenza [Hz] - logscale')
plt.ylabel('Tensione [V]')
plt.xscale('log')
plt.legend()

plt.subplot(1,2,2)
plt.title('Fase')
plt.plot(freqs_1, V_l_1, 'o--', label='$R = 997 \Omega$', color='r', lw=1, ms=2)
plt.plot(freqs_2, V_l_2, 'o--', label='$R = 4021 \Omega$', color='b', lw=1, ms=2)
plt.xlabel('Frequenza [Hz] - logscale')
plt.ylabel('Tensione [V]')
plt.xscale('log')
plt.legend()
plt.savefig('imgs/RLC/confronto_res.pdf', bbox_inches='tight')
```
